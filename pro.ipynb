# Import requets 

```python
import requests 
```

# Users Data

## Check Rest Api Response 

```python
base_url = "https://gorest.co.in"
end_point="/public/v2/users"
response = requests.get(base_url+end_point , params={'page':1 , 'per_page':10})
response
```

## Collect Desired Data 

```python
data = response.json()
data 
```

```python
type(data)
```

```python
data[0]
```

## Convert data from list into CSV 

```python
import csv 
with open('users_data.csv' , mode='w' , newline='' ) as file : 
    # Add Columns Names 
    writer =csv.DictWriter(f= file , fieldnames=['id' , 'name' , 'email' , 'gender' , 'status'])
    writer.writeheader()
    for user in data : 
        id= user['id']
        name = user['name']
        email= user['email']
        gender= user['gender']
        status = user['status']
        writer.writerow({
            'id': id ,
            'name':name , 
            'email':email , 
            'gender': gender, 
            'status' :status
        })
```

# Corona Data

## Check Rest API Response

```python
response = requests.get(url= 'https://coronavirus.m.pipedream.net/')
response
```

## Collect Desired Data

```python
data = response.json()
data
```

```python
data = response.json()['rawData']
data
```

```python
with open('Corona.csv', mode= 'w', newline= '') as file :

    # Add Columns Names
    writer = csv.DictWriter(file, ['Country_Region', 'Confirmed', 'Deaths', 'Recovered'])
    writer.writeheader()

    # Add Rows
    for case_dict in data:

        Country_Region = case_dict['Country_Region']
        Confirmed = case_dict['Confirmed']
        Deaths = case_dict['Deaths']
        Recovered = case_dict['Recovered']

        writer.writerow({'Country_Region' : Country_Region,
                         'Confirmed' : Confirmed,
                         'Deaths' : Deaths,
                         'Recovered' : Recovered})
```
